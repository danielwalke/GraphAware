```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_membership(data, centroids, m):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    membership = 1 / np.sum((distances[:, :, np.newaxis] / distances[:, np.newaxis, :]) ** (2 / (m - 1)), axis=2)
    return membership

def update_centroids(data, membership, m):
    centroids = np.sum((membership[:, :, np.newaxis] ** m) * data[:, np.newaxis], axis=0) / np.sum(membership ** m, axis=0)[:, np.newaxis]
    return centroids

def plot_clusters(data, centroids, membership, iteration):
    plt.figure()
    colors = ['r', 'g', 'b', 'c', 'm']
    
    for i in range(centroids.shape[0]):
        cluster_points = data[np.argmax(membership, axis=1) == i]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[i], label=f'Cluster {i + 1}')
    
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='k', label='Centroids')
    plt.title(f'Iteration {iteration}')
    plt.legend()
    plt.show()

def fuzzy_c_means(data, initial_centroids, m, max_iterations=10, tolerance=1e-4):
    centroids = initial_centroids.copy()
    
    for iteration in range(max_iterations):
        membership = calculate_membership(data, centroids, m)
        new_centroids = update_centroids(data, membership, m)
        
        if np.linalg.norm(new_centroids - centroids) < tolerance:
            break
        
        centroids = new_centroids
        plot_clusters(data, centroids, membership, iteration + 1)
    
    return np.argmax(membership, axis=1)

# Given data points
data_points = np.array([[1, 2], [3, 2], [5, 2], [1, 1], [5, 1]])

# Initial centroids
initial_centroids = np.array([[2.75, 3], [3.25, 0]])

# Fuzzy C-means parameters
fuzziness_parameter = 2

# Perform Fuzzy C-means clustering
cluster_assignments = fuzzy_c_means(data_points, initial_centroids, fuzziness_parameter)

print("Final Cluster Assignments:", cluster_assignments)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def update_membership_degree(data, centroids, m):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    membership = 1 / distances**(2 / (m - 1))
    membership /= np.sum(membership, axis=1, keepdims=True)
    return membership

def update_centroids(data, membership, m):
    centroids = np.sum((membership ** m)[:, :, np.newaxis] * data[:, np.newaxis], axis=0) / np.sum(membership ** m, axis=0)[:, np.newaxis]
    return centroids

def fuzzy_c_means(data, initial_centroids, m, max_iters=100, tol=1e-4):
    centroids = np.array(initial_centroids)
    for i in range(max_iters):
        old_centroids = centroids.copy()
        membership = update_membership_degree(data, centroids, m)
        centroids = update_centroids(data, membership, m)
        
        # Visualization for each iteration
        visualize_clusters(data, membership, centroids, iteration=i)
        
        if np.linalg.norm(centroids - old_centroids) < tol:
            break
    return centroids, membership

def visualize_clusters(data, membership, centroids, iteration):
    colors = np.argmax(membership, axis=1)
    plt.scatter(data[:, 0], data[:, 1], c=colors, cmap='viridis')
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
    plt.title(f'Fuzzy C-Means Clustering - Iteration {iteration}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    data = np.array([[1, 2], [3, 2], [5, 2], [1, 1], [5, 1]])
    initial_centroids = np.array([[2.75, 3], [3.25, 0]])
    m = 2  # Fuzziness parameter

    centroids, membership = fuzzy_c_means(data, initial_centroids, m)
```
